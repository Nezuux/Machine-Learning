````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import HTML, display

# Charger le dataset
url = "https://raw.githubusercontent.com/IBM/telco-customer-churn-on-icp4d/master/data/Telco-Customer-Churn.csv"
df = pd.read_csv(url)

# Nettoyage de base
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
df = df.dropna()  # supprimer les quelques lignes avec TotalCharges vide
df = df.drop('customerID', axis=1)  # identifiant unique, pas une feature

# Encoder le label : Churn -> 1 (parti) / 0 (reste)
df['Churn'] = (df['Churn'] == 'Yes').astype(int)

# Encoder les variables categorielles
df_encoded = pd.get_dummies(df, drop_first=True)

target_col = 'Churn'
feature_names = [c for c in df_encoded.columns if c != target_col]

print(f"Dataset charge : {df.shape[0]} exemples, {df_encoded.shape[1]} colonnes apres encodage")
print(f"Taux de churn : {df['Churn'].mean()*100:.1f}%")
````

<p style="text-align:left; font-size:16px; color:#131fcf"><span>&#128421; <b>Visualiser la distribution du label.</b> L'objectif est de comprendre le déséquilibre entre les classes AVANT de modéliser.<br><br>(1) Utiliser la méthode <b>value_counts</b> de pandas sur la colonne cible pour compter les exemples de chaque classe<br>(2) Tracer un barplot ou un histogramme pour visualiser ce déséquilibre<br>(3) Calculer le ratio entre les deux classes : si une classe représente plus de 70% des exemples, le modèle risque de toujours prédire la classe majoritaire<br><br><em>En classification, le déséquilibre de classes est l'un des pièges les plus courants. Un modèle qui prédit toujours "non" sur un dataset a 80% de "non" aura 80% d'accuracy sans rien avoir appris. C'est pour ca qu'on regarde le recall et l'AUC, pas juste l'accuracy.</em></span></p>

````python
counts = df['Churn'].value_counts().sort_index()
print(counts)
print(f"Ratio classe majoritaire : {counts.max()/counts.sum()*100:.1f}%")
plt.figure(figsize=(5,4))
plt.bar(['Reste (0)', 'Part (1)'], counts.values, color=['steelblue', 'tomato'])
plt.title('Distribution du label Churn')
plt.ylabel('Nombre de clients')
plt.show()
````

<p style="text-align:left; font-size:16px; color:#7a0f43"><span>&#10067; <b>Lire la distribution.</b> La distribution de tenure (durée en tant que client) est-elle uniforme ? Y a-t-il un pic à gauche (clients très récents) ? Quel lien peut-on faire entre l'ancienneté et la probabilité de churn ?</span></p>

````python
# La distribution de tenure n'est pas uniforme.
# On observe un pic à gauche (beaucoup de clients très récents, tenure < 10 mois) et un autre à droite (clients très anciens, tenure > 60 mois).
# Les clients récents ont une probabilité de churn bien plus élevée : ils n'ont pas encore d'engagement fort,
# sont souvent en contrat mensuel, et n'ont pas encore développé d'habitude avec le service.
# Plus l'ancienneté augmente, plus la fidélité s'installe → le churn diminue.

plt.figure(figsize=(8, 4))
plt.hist(df[df['Churn']==0]['tenure'], bins=30, alpha=0.5, label='Reste', color='steelblue')
plt.hist(df[df['Churn']==1]['tenure'], bins=30, alpha=0.5, label='Part', color='tomato')
plt.xlabel('Tenure (mois)')
plt.ylabel('Nombre de clients')
plt.title('Distribution de tenure par statut Churn')
plt.legend()
plt.show()
````

<p style="text-align:left; font-size:16px; color:#131fcf"><span>&#128421; <b>Découper en train et test.</b> L'objectif est de garder 20% des données de côté pour évaluer le modèle sur des données qu'il n'a JAMAIS vues.<br><br>(1) Utiliser la fonction <b>train_test_split</b> de sklearn pour découper X et y en 4 variables : X_train, X_test, y_train, y_test<br>(2) Fixer le paramètre test_size a 0.2 (20% pour le test) et random_state a 42 (pour la reproductibilité)<br>(3) Afficher les tailles des 4 résultats pour vérifier le découpage<br><br><em>Pourquoi 80/20 ? <p style="text-align:left; font-size:16px; color:#131fcf"><span>&#128421; <b>Construire votre premier modèle.</b> L'objectif est d'enchaîner le preprocessing et la modélisation dans un Pipeline sklearn pour éviter tout risque de data leakage.<br><br>(1) Importer la classe <b>Pipeline</b> de sklearn et le modèle de base (régression linéaire ou logistique selon votre tache)<br>(2) Créer un Pipeline avec deux étapes nommées : un StandardScaler en première position pour normaliser les features, puis le modèle en deuxième position<br>(3) Entraîner le Pipeline sur le jeu d'entraînement avec la méthode fit — en une seule ligne, le scaler apprend les statistiques ET le modèle apprend les poids<br>(4) Stocker le Pipeline dans une variable que vous réutiliserez pour les prédictions<br><br><em>Le Pipeline garantit que le scaling est appris UNIQUEMENT sur le train set, jamais sur le test set. <p style="text-align:left; font-size:16px; color:#131fcf"><span>&#128421; <b>Expérimenter : combien d'arbres faut-il ?</b> Le paramètre <code>n_estimators</code> contrôle le nombre d'arbres. Plus il y en a, plus le modèle est potentiellement précis — mais plus l'entraînement est long. Re-entraîner le RandomForestClassifier avec <code>n_estimators=200</code> puis <code>n_estimators=500</code>. Pour chacun, noter le score AUC et le temps d'exécution.
````python
import time

results_n_est = []
for n in [100, 200, 500]:
    t0 = time.time()
    rf_temp = Pipeline([
        ('scaler', StandardScaler()),
        ('regressor', RandomForestClassifier(n_estimators=n, random_state=42))
    ])
    rf_temp.fit(X_train, y_train)
    elapsed = time.time() - t0
    y_proba_temp = rf_temp.predict_proba(X_test)[:, 1]
    auc_temp = roc_auc_score(y_test, y_proba_temp)
    results_n_est.append({'n_estimators': n, 'AUC': auc_temp, 'Temps (s)': round(elapsed, 2)})
    print(f"n_estimators={n:>3} | AUC = {auc_temp:.4f} | Temps = {elapsed:.2f}s")

print("\nRésumé :")
print(pd.DataFrame(results_n_est).to_string(index=False))
````

<p style="text-align:left; font-size:16px; color:#7a0f43"><span>&#10067; <b>Le compromis performance / coût.</b> (1) Notez le AUC obtenu avec 100, 200 et 500 arbres. Le gain est-il proportionnel ? (2) Si passer de 200 a 500 arbres ne gagne que +0.001 de AUC mais double le temps de calcul, est-ce que ca vaut le coup ? (3) En entreprise, chaque heure de calcul GPU coûte de l'argent. Pour le détecter : comparer le score sur le train set et le score sur le test set. Si le train est a 0.95 et le test a 0.83, l'écart est significatif — le modèle a "appris par coeur" les données d'entraînement. Un écart &gt; 0.10 est un signal d'alerte.</span></p>

<p style="text-align:left; font-size:16px; color:#131fcf"><span>&#128421; <b>Définir quelles combinaisons tester.</b> Le GridSearchCV va essayer systématiquement toutes les combinaisons d'hyperparametres qu'on lui donne. Définir un <code>param_grid</code> (dictionnaire) avec au moins 2 valeurs pour chaque hyperparametre clé du GradientBoostingClassifier : <code>n_estimators</code> (nombre d'arbres : essayez 100, 200, 300), <code>max_depth</code> (profondeur max : essayez 3, 5, 7 — plus profond = plus de risque d'overfitting), <code>learning_rate</code> (vitesse d'apprentissage : essayez 0.05, 0.1 — plus petit = plus stable mais plus lent). Important : dans un Pipeline, les noms sont préfixes : <code>'regressor__n_estimators'</code>.</span></p>

````python
from sklearn.model_selection import GridSearchCV

model_to_optimize = Pipeline([
    ('scaler', StandardScaler()),
    ('regressor', GradientBoostingClassifier(random_state=42))
])

param_grid = {
    'regressor__n_estimators': [100, 200, 300],
    'regressor__max_depth': [3, 5, 7],
    'regressor__learning_rate': [0.05, 0.1, 0.2],
}
````

<p style="text-align:left; font-size:16px; color:#131fcf"><span>&#128421; <b>Lancer la recherche automatique.</b> L'objectif est de tester toutes les combinaisons d'hyperparametres et de trouver la meilleure par validation croisée.<br><br>(1) Créer un objet GridSearchCV en lui passant le modèle a optimiser, le dictionnaire de paramètres, le nombre de folds (5), et la métrique de scoring (AUC pour la classification)<br>(2) Utiliser n_jobs=-1 pour paralléliser sur tous les coeurs du processeur<br>(3) Lancer la recherche avec la méthode fit — cela peut prendre quelques minutes<br>(4) Afficher les meilleurs paramètres trouves et le meilleur score de cross-validation<br><br><em>Le GridSearchCV teste systématiquement TOUTES les combinaisons. Si vous avez 3 valeurs pour 3 paramètres, ca fait 27 combinaisons x 5 folds = 135 entraînements. C'est pour ca que c'est plus lent — mais le résultat est bien plus fiable qu'un réglage manuel.</em></span></p>

````python
grid_search = GridSearchCV(
    model_to_optimize, 
    param_grid, 
    cv=5, 
    scoring='roc_auc', 
    verbose=1, 
    n_jobs=-1
)
grid_search.fit(X_train, y_train)

print(f"\nMeilleurs paramètres : {grid_search.best_params_}")
print(f"Meilleur score CV (AUC) : {grid_search.best_score_:.4f}")

best_model = grid_search.best_estimator_
````

Comparer : (1) au score avant optimisation (GradientBoostingClassifier par défaut), (2) a la baseline (régression logistique). <p style="text-align:left; font-size:16px; color:#7a0f43"><span>&#10067; <b>Coût des erreurs.</b> Pour un responsable marketing, quel type d'erreur est le plus coûteux : un faux négatif (rater un vrai cas) ou un faux positif (alerter a tort) ? Le coût est-il symétrique ? Comment ajusteriez-vous le seuil de décision en conséquence ?</span></p>

````python
# Le faux négatif est bien plus coûteux pour un responsable marketing :
# - Faux négatif : un client à risque passe inaperçu → il résilie → perte de revenu + coût de réacquisition (~5-25x le coût de rétention)
# - Faux positif : un client fidèle reçoit un appel de rétention inutile → coût faible (temps d'un conseiller, ~5-10€)
# Les coûts ne sont PAS symétriques : perdre un client coûte 100-500€, un appel inutile coûte 5-10€.
# 
# Pour ajuster : on peut BAISSER le seuil de décision (ex: prédire "churn" dès 0.3 au lieu de 0.5)
# → on détecte plus de vrais positifs (recall ↑) mais on génère aussi plus de faux positifs (précision ↓)
# C'est un compromis acceptable vu l'asymétrie des coûts.
print("Coût faux négatif >> Coût faux positif → privilégier le recall, baisser le seuil.")
````
